```python
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
from keras.applications import InceptionV3
```

```python
IMAGE_SIZE = [224,224]
train_path = '/content/drive/MyDrive/card dataset/dataset/train'
valid_path = '/content/drive/MyDrive/card dataset/dataset/valid'
```

```python
inc = InceptionV3(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
for layer in inc.layers:
    layer.trainable = False
```

```python
folders = glob("/content/drive/MyDrive/card dataset/dataset/train/*")
folders
```

```python
len(folders)
```

```python
x = Flatten()(inc.output)
```

```python
prediction = Dense(len(folders), activation='softmax')(x)
model = Model(inputs=inc.input, outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/card dataset/dataset/train',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory('/content/drive/MyDrive/card dataset/dataset/valid',
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
cnn_model = model.fit_generator(
  training_set,
  validation_data=test_set,
  epochs=25,
  steps_per_epoch=len(training_set),
  validation_steps=len(test_set)
)
```

```python
from tensorflow.keras.models import load_model
model.save('card_pred.h5')
```
